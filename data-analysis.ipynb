```python
#Merging data - step 1, doing in two steps

#Merging Wild Type Data
df_W = pd.concat([
                  df_ET1,
                  df_ET2,
                  df_ET5,
                  df_ET6],
                  ignore_index = True)


#Merging delP168/S144A data
df_M = pd.concat([
                  df_ET3,
                  df_ET4,
                  df_ET7,
                  df_ET8],
                  ignore_index = True)
df_M
```

```python
# Converting data to long form

df_W = pd.melt(df_W, id_vars = ['Time [s]', 'replicate'])

df_M = pd.melt(df_M, id_vars = ['Time [s]', 'replicate'])

#Setting experiment flags

df_M['experiment'] = "DelP168_S144A"

df_W['experiment'] = "WT"

df_M
```

```python
#Merging datasets into one.

df = pd.concat([df_W,
                df_M],
                ignore_index = True)

df
```

```python
# quick look at data
sns.scatterplot(x='Time [s]',
                y='value',
                hue='variable',
                data=df
               )
# bands on top are controls
```

```python
# remove positive controls
df_control = df[df.variable.str.contains('AEDANS')]
df = df[~df.variable.str.contains('AEDANS')]
print(f'{len(df_control)} positive control rows removed')
print(f'{len(df)} rows remaining')
```

```python
# function to make and save plots
def plt_key(exp):
    # variable 'key' is string of experiment
    print(f"key: {exp}")
    fig, ax1 = plt.subplots(1,1)
    g = sns.lineplot(
                 data=df.loc[df["experiment"] == exp],
                 x="Time [s]",
                 y="value",
                 hue="variable"
                )
    g.set_title(f'SARS-CoV-2 Mpro {exp} Ensitrelvir Titration Kinetics')
    g.set_xlabel(f'Time (s)')
    g.set_ylabel(f'Relative Fluorescence')
    g.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.savefig(f'images/plot-{exp}.png', dpi=300)
    plt.savefig(f'images/plot-{exp}.svg', dpi=300)
    plt.show()
    """
    sns.scatterplot(x='Time [s]',
                y='value',
                hue='variable',
                data=df.loc[df["experiment"] == exp]
               )
    """
```
